```python
#webscrapper + text analyzer

import requests
from bs4 import BeautifulSoup as bs
from textblob import TextBlob

user="realDonaldTrump" #chosen use
htag="realDonaldTrump"

def twitter_analyser(topic):

    if topic==0:
        site=requests.get("https://twitter.com/"+user)
        site_con=site.content
        soup=bs(site_con,"html.parser")
        for link in soup.findAll('p',{"class":['TweetTextSize','TweetTextSize--normal','js-tweet-text tweet-text']}):
            analysis = TextBlob(link.text)
            print(link.text+"\n\n\n")
            print(analysis.sentiment)
            
    if topic==1:# make it dynamic   !!WIP!!
        site=requests.get("https://twitter.com/hashtag/"+htag)
        site_con=site.content
        soup=bs(site_con,"html.parser")
        for link in soup.findAll('li',{"class":['js-stream-item','TweetTextSize--normal','js-tweet-text tweet-text']}):
            #analysis = TextBlob(link.text)
            #print(link.text+"\n\n\n")
            #print(analysis.sentiment)
            print(link)

    
    #js-stream-item stream-item stream-item

twitter_analyser(0)# if input is 0 it will analyzer the tweet .... if input is 1 it will show all
```
